```python
#Libraries
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from pandas_ods_reader import read_ods

```

```python
#Importing the dataset
file_path = "dataset/CLIMO_ACCRA_2020.ods"
df = read_ods(file_path)
#df = df.dropna()
#filling missing values with 0.0
df = df.fillna(0.0)
df.head(30)
```

```python
#Parameters needed for rainfall
rainfall = df[df['ELEMENTS'] == 'RR']
rainfall_month = rainfall['Month'].values
rain_monthly_avg = []
month_rain = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Oct','Nov','Dec']

#Parameters needed for wind speed
wind_speed = df[df['ELEMENTS'] == 'Kts']
wind_month = wind_speed['Month'].values
wind_monthly_avg = []
month_wind = ['Feb','Mar','Apr','May','Jun','Jul','Aug','Nov','Dec']

```

```python
def monthly_avg(key):

    parameter = df[df['ELEMENTS'] == key] 
    i = 0
    if key == 'RR':
        while i < len(rainfall_month): #looping through the months
            values = parameter[parameter['Month']== rainfall_month[i]].loc[0:,'DAY1':'DAY31'].values #extracting the values of the day1 to day31
            rain_monthly_avg.append(np.average(values)) #summing the values of the day1 to day31
            i = i + 1


    elif key == 'Kts':
        while i < len(wind_month): #looping through the months
            values = parameter[parameter['Month']== wind_month[i]].loc[0:,'DAY1':'DAY31'].values 
            wind_monthly_avg.append(np.average(values))  
            i = i + 1
monthly_avg('RR')
monthly_avg('Kts')
print(rain_monthly_avg)
print(wind_monthly_avg)
```

```python
#Plotting the graph for rainfall
plt.figure(figsize = (13,8))
plt.bar(month_rain, rain_monthly_avg, width= 0.8, color='#C2CE2B')
plt.ylabel('AMOUNT OF RAINFALL / mm', labelpad=25)
plt.xlabel('MONTHS', labelpad=15)
plt.grid()
#Using the detailed element for graph title
plt.title('AVERAGE MONTHLY RAINFALL BAR GRAPH FOR 2020 AT ACCRA', pad=25)

plt.savefig('images/rainfall_bar.png')
plt.show() 





```

```python
plt.figure(figsize = (13,8))

plt.plot(month_rain, rain_monthly_avg,color='#C2CE2B', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.ylabel('AMOUNT OF RAINFALL / mm', labelpad=25)
plt.xlabel('MONTHS', labelpad=15)
plt.grid()
#Using the detailed element for graph title
plt.title('AVERAGE MONTHLY RAINFALL LINE GRAPH FOR 2020 AT ACCRA', pad=25)

plt.savefig('images/rainfall_line.png')
```

```python
#Plotting the graph for rainfall
plt.figure(figsize = (13,8))
plt.bar(month_rain, rain_monthly_avg, width= 0.8, color='#0989FA')
plt.ylabel('WIND SPEED / kts', labelpad=25)
plt.xlabel('MONTHS', labelpad=15)
plt.grid()
#Using the detailed element for graph title
plt.title('AVERAGE MONTHLY WIND SPEED BAR GRAPH FOR 2020 AT ACCRA', pad=25)

plt.savefig('images/wind_bar.png')
plt.show() 
```

```python
plt.figure(figsize = (13,8))

plt.plot(month_wind, wind_monthly_avg,color='#0989FA', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.ylabel('WIND SPEED / kts', labelpad=25)
plt.xlabel('MONTHS', labelpad=15)
plt.grid()
#Using the detailed element for graph title
plt.title('AVERAGE MONTHLY WIND SPEED LINE GRAPH FOR 2020 AT ACCRA', pad=25)

plt.savefig('images/wind_line.png')
plt.show()
```
